```python
import pandas as pd

# Load the dataset
df = pd.read_csv('origins.csv')
```

```python
from sklearn.preprocessing import StandardScaler, LabelEncoder
# Step 1: Drop unnecessary columns
df = df.drop(columns=['Track ID', 'Track Name', 'Album Name', 'Artist Name(s)', 'Release Date', 'Added By', 'Added At', 'Record Label'])

# Step 2: Handle missing values in the 'Genres' column
df['Genres'] = df['Genres'].fillna('')  # Replace NaN with empty string

# Step 3: Encode the 'Genres' column (multiple genres can be separated by commas)
df['Genres'] = df['Genres'].apply(lambda x: x.split(',')[0] if x else '')  # Just keep the first genre
label_encoder = LabelEncoder()
df['Genres'] = label_encoder.fit_transform(df['Genres'])

# Step 4: Check the cleaned dataframe
df.head()
```

```python
# Example: Binary classification based on popularity
df['Popular'] = (df['Popularity'] > 50).astype(int)  # Popular = 1 if Popularity > 50, else 0

# Clean column names
df.columns = df.columns.str.strip()

# Drop unnecessary columns (if any)
X = df.drop(['Popularity', 'Popular'], axis=1)  # Features
y = df['Popular']  # Target

# Normalize numerical features
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)
```

```python
# Check class distribution
print(df['Popular'].value_counts(normalize=True))
```

```python
from imblearn.over_sampling import SMOTE

# Initialize SMOTE
smote = SMOTE(random_state=42)

# Apply SMOTE to training data
X_resampled, y_resampled = smote.fit_resample(X_train, y_train)

# Check class distribution after resampling
print(pd.Series(y_resampled).value_counts(normalize=True))
```

```python
from sklearn.linear_model import LogisticRegression

# Initialize and train the model
lr = LogisticRegression(random_state=42)
lr.fit(X_resampled, y_resampled)

# Predict on the test set
y_pred_lr = lr.predict(X_test)
```

```python
from sklearn.ensemble import RandomForestClassifier

# Initialize and train the model
rf = RandomForestClassifier(random_state=42)
rf.fit(X_resampled, y_resampled)

# Predict on the test set
y_pred_rf = rf.predict(X_test)
```

```python
from sklearn.svm import SVC

# Initialize and train the model
svm = SVC(random_state=42)
svm.fit(X_resampled, y_resampled)

# Predict on the test set
y_pred_svm = svm.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

# Example for Logistic Regression
print("Logistic Regression:")
print(f"Accuracy: {accuracy_score(y_test, y_pred_lr):.4f}")
print(f"Precision: {precision_score(y_test, y_pred_lr):.4f}")
print(f"Recall: {recall_score(y_test, y_pred_lr):.4f}")
print(f"F1-Score: {f1_score(y_test, y_pred_lr):.4f}")
```

```python
results = {
    'Model': ['Logistic Regression', 'Random Forest', 'SVM'],
    'Accuracy': [accuracy_score(y_test, y_pred_lr),
                 accuracy_score(y_test, y_pred_rf),
                 accuracy_score(y_test, y_pred_svm)],
    'Precision': [precision_score(y_test, y_pred_lr),
                  precision_score(y_test, y_pred_rf),
                  precision_score(y_test, y_pred_svm)],
    'Recall': [recall_score(y_test, y_pred_lr),
               recall_score(y_test, y_pred_rf),
               recall_score(y_test, y_pred_svm)],
    'F1-Score': [f1_score(y_test, y_pred_lr),
                 f1_score(y_test, y_pred_rf),
                 f1_score(y_test, y_pred_svm)]
}

import pandas as pd
results_df = pd.DataFrame(results)
print(results_df)
```
